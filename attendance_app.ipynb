```python
#################################################################
def assure_path_exists(path):
    dir = os.path.dirname(path)
    if not os.path.exists(dir):
        os.makedirs(dir)
#################################################################
def tick():
    time_string = time.strftime('%H:%M:%S')
    clock.config(text=time_string)
    clock.after(200,tick)
#################################################################
def contact():
    mess._show(title='Contact us', message="Please contact us on :'prajapatramesh520@gmail.com' ")
#################################################################
def check_haarcascadefile():
    exists = os.path.isfile("haarcascade_frontalface_default.xml")
    if exists:
        pass
    else:
        mess._show(title='Some file missing', message='Please contact us for help')
        window.destroy()
        
###################################################################

def back_to_main(window):
    """Destroy the current window and return to the main menu."""
    window.destroy()
    main_menu()
    
#################################################################

def save_pass():
    assure_path_exists("TrainingImageLabel/")
    exists1 = os.path.isfile("TrainingImageLabel\psd.txt")
    if exists1:
        tf = open("TrainingImageLabel\psd.txt", "r")
        key = tf.read()
    else:
        master.destroy()
        new_pas = tsd.askstring('Old Password not found', 'Please enter a new password below',show='*')
        if new_pas == None:
            mess._show(title='No Password Entered', message='Password not set!! Please try again')
        else:
            tf = open("TrainingImageLabel\psd.txt", "w")
            tf.write(new_pas)
            mess._show(title='Password Registered', message='New password was registered successfully!!')
            return
    op = (old.get())
    newp= (new.get())
    nnewp = (nnew.get())
    if (op == key):
        if(newp == nnewp):
            txf = open("TrainingImageLabel\psd.txt", "w")
            txf.write(newp)
        else:
            mess._show(title='Error', message='Confirm new password again!!!')
            return
    else:
        mess._show(title='Wrong Password', message='Please enter correct old password.')
        return
    mess._show(title='Password Changed', message='Password changed successfully!!')
    master.destroy()
#################################################################

def change_pass():
    global master
    master = tk.Tk()
    master.geometry("400x160")
    master.resizable(False,False)
    master.geometry(f"+{550}+{350}")
    master.title("Change Password")
    master.configure(background="white")
    lbl4 = tk.Label(master,text=' Enter Old Password',bg='white',font=('times', 12, ' bold '))
    lbl4.place(x=10,y=10)
    global old
    old=tk.Entry(master,width=25 ,fg="black",relief='solid',font=('times', 12, ' bold '),show='*')
    old.place(x=180,y=10)
    lbl5 = tk.Label(master, text=' Enter New Password', bg='white', font=('times', 12, ' bold '))
    lbl5.place(x=10, y=45)
    global new
    new = tk.Entry(master, width=25, fg="black",relief='solid', font=('times', 12, ' bold '),show='*')
    new.place(x=180, y=45)
    lbl6 = tk.Label(master, text='Confirm New Password', bg='white', font=('times', 12, ' bold '))
    lbl6.place(x=10, y=80)
    global nnew
    nnew = tk.Entry(master, width=25, fg="black", relief='solid',font=('times', 12, ' bold '),show='*')
    nnew.place(x=180, y=80)
    cancel=tk.Button(master,text="Cancel", command=master.destroy ,fg="black" ,bg="red" ,height=1,width=25 , activebackground = "white" ,font=('times', 10, ' bold '))
    cancel.place(x=200, y=120)
    save1 = tk.Button(master, text="Save", command=save_pass, fg="black", bg="#3ece48",height = 1,width=25, activebackground="white", font=('times', 10, ' bold '))
    save1.place(x=10, y=120)
    master.mainloop()
    
    
#################################################################

def id_validation(Id):
    registered_students = []
    with open("StudentDetails\StudentDetails.csv", 'r') as csvFile1:
        reader1 = csv.reader(csvFile1)
        lst = list(reader1)
        i = 1
        for row in lst:
            i = i+1
            if(i % 2 == 0):
                registered_students.append(row[2])

    if Id in registered_students:
        mess._show(title='ID Already Exist...', message='Entered ID already exists...')
        return 1

#################################################################

def psw():
    assure_path_exists("TrainingImageLabel/")
    exists1 = os.path.isfile("TrainingImageLabel\psd.txt")
    valid  = id_validation(Idpsw)
    if valid:
        
        return
    
    exists = os.path.isfile("StudentDetails\StudentDetails.csv")
    if exists:
        registered_students = []
        with open("StudentDetails\StudentDetails.csv", 'r') as csvFile1:
            reader1 = csv.reader(csvFile1)
            lst = list(reader1)
            i = 1
            for row in lst:
                i = i+1
                if(i % 2 == 0):
                    registered_students.append(row[2])
        if Idpsw in registered_students:
            mess._show(title='ID Already Exist...', message='Entered ID already exists...')
            return 
    
    if exists1:
        tf = open("TrainingImageLabel\psd.txt", "r")
        key = tf.read()
    else:
        new_pas = tsd.askstring('Old Password not found', 'Please enter a new password below',show='*')
        if new_pas == None:
            mess._show(title='No Password Entered', message='Password not set!! Please try again')
        else:
            tf = open("TrainingImageLabel\psd.txt", "w")
            tf.write(new_pas)
            mess._show(title='Password Registered', message='New password was registered successfully!!')
            return
    password = tsd.askstring('Password', 'Enter Password', show='*')
    if (password == key):
        TrainImages()
    elif (password == None):
        pass
    else:
        mess._show(title='Wrong Password', message='You have entered wrong password')
#################################################################


    
    
#################################################################

def clear():
    txt.delete(0, 'end')
    res = "NOTE:- 1)Take Images >>> 2)Save Profile"
    message1.configure(text=res)
    
def clear2():
    txt2.delete(0, 'end')
    res = "NOTE:- 1)Take Images >>> 2)Save Profile"
    message1.configure(text=res)
#################################################################

def TakeImages():
    check_haarcascadefile()
    columns = ['SERIAL NO.', '', 'ID', '', 'NAME']
    assure_path_exists("StudentDetails/")
    assure_path_exists("TrainingImage/")
    global serial
    serial = 0
    exists = os.path.isfile("StudentDetails\StudentDetails.csv")
    
    if exists:
        with open("StudentDetails\StudentDetails.csv", 'r') as csvFile1:
            reader1 = csv.reader(csvFile1)
            for l in reader1:
                serial = serial + 1
        serial = (serial // 2)
        csvFile1.close()
        
    else:
        with open("StudentDetails\StudentDetails.csv", 'a+') as csvFile1:
            writer = csv.writer(csvFile1)
            writer.writerow(columns)
            serial = 1
        csvFile1.close()
    Id = (txt.get())
    name = (txt2.get())
    global Idpsw
    Idpsw = Id
    valid  = id_validation(Id)
    if valid:
        return
    if ((name.isalpha()) or (' ' in name)):
        cam = cv2.VideoCapture(0)
        harcascadePath = "haarcascade_frontalface_default.xml"
        detector = cv2.CascadeClassifier(harcascadePath)
        sampleNum = 0
        while (True):
            ret, img = cam.read()
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = detector.detectMultiScale(gray, 1.3, 5)
            for (x, y, w, h) in faces:
                cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
                # incrementing sample number
                sampleNum = sampleNum + 1
                # saving the captured face in the dataset folder TrainingImage
                cv2.imwrite("TrainingImage\ " + name + "." + str(serial) + "." + Id + '.' + str(sampleNum) + ".jpg", gray[y:y + h, x:x + w])
                # display the frame
                cv2.imshow('Taking Images' , img)
            # wait for 100 miliseconds
            if cv2.waitKey(100) & 0xFF == ord('q'):
                break
            # break if the sample number is morethan 100
            elif sampleNum > 100:
                break
        cam.release()
        cv2.destroyAllWindows()
        res = "Images Taken for ID : " + Id
        row = [serial, '', Id, '', name]
        with open('StudentDetails\StudentDetails.csv', 'a+') as csvFile:
            writer = csv.writer(csvFile)
            writer.writerow(row)
        csvFile.close()
        message1.configure(text=res)
    else:
        if (name.isalpha() == False):
            res = "Enter Correct name"
            message.configure(text=res)
#################################################################


def TrainImages():
    check_haarcascadefile()
    assure_path_exists("TrainingImageLabel/")
    recognizer = cv2.face_LBPHFaceRecognizer.create()
    harcascadePath = "haarcascade_frontalface_default.xml"
    detector = cv2.CascadeClassifier(harcascadePath)
    faces, ID = getImagesAndLabels("TrainingImage")
    try:
        recognizer.train(faces, np.array(ID))
    except:
        mess._show(title='No Registrations', message='Please Register someone first!!!')
        return
    recognizer.save("TrainingImageLabel\Trainner.yml")
    res = 0
    exists = os.path.isfile("StudentDetails\StudentDetails.csv")
    if exists:
        with open("StudentDetails\StudentDetails.csv", 'r') as csvFile1:
            reader1 = csv.reader(csvFile1)
            for l in reader1:
                res = res + 1
        res = (res // 2) - 1
        csvFile1.close()
    else:
        res = 0
    message.configure(text='Total Registrations till now : '+str(res),width="30", bg="#ded7f1")
    message.place(x=120,y=500)
    
#################################################################

def getImagesAndLabels(path):
    # get the path of all the files in the folder
    imagePaths = [os.path.join(path, f) for f in os.listdir(path)]
    # create empth face list
    faces = []
    # create empty ID list
    Ids = []
    # now looping through all the image paths and loading the Ids and the images
    for imagePath in imagePaths:
        # loading the image and converting it to gray scale
        pilImage = Image.open(imagePath).convert('L')
        # Now we are converting the PIL image into numpy array
        imageNp = np.array(pilImage, 'uint8')
        # getting the Id from the image
        ID = int(os.path.split(imagePath)[-1].split(".")[1])
        # extract the face from the training image sample
        faces.append(imageNp)
        Ids.append(ID)
    return faces, Ids
#################################################################

def TrackImages():
    check_haarcascadefile()
    assure_path_exists("Attendance/")
    assure_path_exists("StudentDetails/")
    for k in tv.get_children():
        tv.delete(k)
    msg = ''
    i = 0
    j = 0
    recognizer = cv2.face.LBPHFaceRecognizer_create() 
    exists3 = os.path.isfile("TrainingImageLabel\Trainner.yml")
    if exists3:
        recognizer.read("TrainingImageLabel\Trainner.yml")
    else:
        mess._show(title='Data Missing', message='Please click on Save Profile to reset data!!')
        return
    harcascadePath = "haarcascade_frontalface_default.xml"
    faceCascade = cv2.CascadeClassifier(harcascadePath)
    font = cv2.FONT_HERSHEY_SIMPLEX
    col_names = ['Id', '', 'Name', '', 'Date', '', 'Time']
    exists1 = os.path.isfile("StudentDetails\StudentDetails.csv")
    if exists1:
        df = pd.read_csv("StudentDetails\StudentDetails.csv")
    else:
        mess._show(title='Details Missing', message='Students details are missing, please check!')
        cam.release()
        cv2.destroyAllWindows()
        window.destroy()
        return
    
    ts = time.time()
    date = datetime.datetime.fromtimestamp(ts).strftime('%d-%m-%Y')
    attendance_file = "Attendance\Attendance_" + date + ".csv"
    exists_attendance = os.path.isfile(attendance_file)
    present_students = []
    if exists_attendance:
        with open(attendance_file, 'r') as csvFile1:
            reader1 = csv.reader(csvFile1)
            for lines in reader1:
                if len(lines) > 0:
                    present_students.append(lines[0])
    # Show attendance on tree view if there is at least one student present on current day
    if exists_attendance:
        with open(attendance_file, 'r') as csvFile1:
            reader1 = csv.reader(csvFile1)
            for lines in reader1:
                if len(lines) > 0:
                    iidd = str(lines[0]) + ' '
                    tv.insert('', 0, text=iidd, values=(str(lines[2]), str(lines[4]), str(lines[6])))
    else:
        tv.insert('', 0, text="No attendance for today", values=("", "", ""))
    attendance_image_folder = "Attendance_Images/" + date + "/"
    assure_path_exists(attendance_image_folder)
    # Open camera
    cam = cv2.VideoCapture(0)
    font = cv2.FONT_HERSHEY_SIMPLEX
    while True:  
        ret, im = cam.read()  
        gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)  
        faces = faceCascade.detectMultiScale(gray, 1.2, 5)  
        for (x, y, w, h) in faces:  
            cv2.rectangle(im, (x, y), (x + w, y + h), (225, 0, 0), 2)  
            serial, conf = recognizer.predict(gray[y:y + h, x:x + w])  
            if (conf < 50):  
                ts = time.time()  
                date = datetime.datetime.fromtimestamp(ts).strftime('%d-%m-%Y')  
                timeStamp = datetime.datetime.fromtimestamp(ts).strftime('%H:%M:%S')  
                aa = df.loc[df['SERIAL NO.'] == serial]['NAME'].values  
                ID = df.loc[df['SERIAL NO.'] == serial]['ID'].values  
                ID = str(ID)  
                ID = ID[1:-1]  
                bb = str(aa)  
                bb = bb[2:-2]  
                attendance = [str(ID), '', bb, '', str(date), '', str(timeStamp)]  
                if ID not in present_students:
                    present_students.append(ID)
                    with open(attendance_file, 'a', newline='') as csvFile1:
                        writer = csv.writer(csvFile1)
                        writer.writerow(attendance)
                    for k in tv.get_children():
                        tv.delete(k)
                    iidd = str(ID) + ' '
                    tv.insert('', 0, text=iidd, values=(str(bb), str(date), str(timeStamp)))
                    # Save the attendance image
                    cv2.imwrite(attendance_image_folder + ID + "." + bb +  ".jpg", gray[y:y + h, x:x + w])
                    cam.release()  
                    cv2.destroyAllWindows()
                    return
                else:
                    mess._show(title='Attendance Already Marked', message='Attendance for ' + bb  + " (ID=" + ID + ")"+ ' has already been marked for today.')  
                    cam.release()  
                    cv2.destroyAllWindows()  
                    return
            else:
                Id = 'Unknown'
                bb = str(Id)
            cv2.putText(im, str(bb), (x, y + h), font, 1, (255, 255, 255), 2)
        cv2.imshow('Taking Attendance', im)
        if (cv2.waitKey(1) == ord('q') or cv2.waitKey(1) == 27):
            cam.release()
            cv2.destroyAllWindows()
            break
        # Refresh the treeview with all attendance for today
        for k in tv.get_children():
            tv.delete(k)
        with open(attendance_file, 'r') as csvFile1:
            reader1 = csv.reader(csvFile1)
            next(reader1)  # Skip header
            for lines in reader1:
                tv.insert('', 0, text=str(lines[0]), values=(str(lines[2]), str(lines[4]), str(lines[6])))
        csvFile1.close()
        

################################################################

def load_attendance():
    ts = time.time()
    date = datetime.datetime.fromtimestamp(ts).strftime('%d-%m-%Y')
    attendance_file = "Attendance\Attendance_" + date + ".csv"
    exists_attendance = os.path.isfile(attendance_file)

    for k in tv.get_children():
        tv.delete(k)

    if exists_attendance:
        with open(attendance_file, 'r') as csvFile1:
            reader1 = csv.reader(csvFile1)
            for lines in reader1:
                if len(lines) > 0:
                    iidd = str(lines[0]) + ' '
                    tv.insert('', 0, text=iidd, values=(str(lines[2]), str(lines[4]), str(lines[6])))
    else:
        tv.insert('', 0, text="No attendance taken till now", values=("", "", ""))


################################################################

def generate_report():
    check_haarcascadefile()
    assure_path_exists("Reports/")
    assure_path_exists("Attendance/")
    password = tsd.askstring('Password', 'Enter Administrator Password', show='*')
    if password is None:
        mess._show(title='No Password Entered', message='Password not set!! Please try again')
    else:
        dir = os.path.dirname("TrainingImageLabel/psd.txt")
        if not os.path.exists(dir):
            mess._show(title='Attendance data not available', message='Attendance data is not available...\nAtleast one attendance is required to generate a report.')
            return
        with open("TrainingImageLabel/psd.txt", "r") as tf:
            key = tf.read()
        if password == key:
            report_window = tk.Toplevel(window)
            report_window.title("Attendance Report")
            report_window.geometry("1920x1080")
            report_window.configure(background='#f0f0f0')

            report_frame = tk.Frame(report_window, bg="#cce7ff")
            report_frame.place(relx=0.5, rely=0.5, anchor='center', relwidth=0.60, relheight=0.78)

#             report_label = tk.Label(report_frame, text="Select Report Type", fg="black", bg="#ded7f1", font=('times', 17, 'bold'))
#             report_label.place(relx=0.5, rely=0.05, anchor='center')
            
            report_label = tk.Label(report_frame, text="Select Report Type", fg="black", font=('times', 17, 'bold'))
            report_label.place(relx=0.05, rely=0.05, relwidth=0.9, relheight=0.1)

            report_type = tk.StringVar()
            report_type.set("All Students")
            report_type_menu = tk.OptionMenu(report_frame, report_type, "All Students", "Particular Student")
            report_type_menu.place(relx=0.5, rely=0.18, anchor='center')

            report_button = tk.Button(report_frame, text="Generate Report", command=lambda: generate_report_details(report_type.get()), 
                                      fg="#660066", bg="#388e3c", width=18, height=1, activebackground="#7a6ab2", font=('times', 15, 'bold'))
            report_button.place(relx=0.5, rely=0.25, anchor='center')

            report_text = tk.Text(report_frame, width=80, height=15)
            report_text.place(relx=0.5, rely=0.5, anchor='center')

            # Positioning the buttons closer to the report_text
            download_button = tk.Button(report_frame, text="Download Report", command=lambda: download_report(report_text.get(1.0, tk.END)), 
                                        fg="#660066", bg="#388e3c", width=18, height=1, activebackground="#7a6ab2", font=('times', 15, 'bold'))
            download_button.place(relx=0.3, rely=0.78, anchor='center')

            back_button = tk.Button(report_frame, text="Back", command=report_window.destroy, 
                                    fg="#660066", bg="#d32f2f", width=18, height=1, activebackground="#7a6ab2", font=('times', 15, 'bold'))
            back_button.place(relx=0.7, rely=0.78, anchor='center')

            def generate_report_details(report_type):
                report_text.delete(1.0, tk.END)
                if report_type == "All Students":
                    all_students_report()
                elif report_type == "Particular Student":
                    particular_student_report()
                    
            def all_students_report():
                for file in os.listdir("Attendance/"):
                    if file.endswith(".csv"):
                        with open("Attendance/" + file, 'r') as csvFile1:
                            reader1 = csv.reader(csvFile1)
                            for lines in reader1:
                                if len(lines) > 0:
                                    report_text.insert(tk.END, f"{lines[0]} - {lines[2]} - {lines[4]} - {lines[6]}\n")

            def particular_student_report():
                student_id = tsd.askstring('Student ID', 'Enter Student ID')
                if student_id is None:
                    mess._show(title='No Student ID Entered', message='Student ID not set!! Please try again')
                else:
                    for file in os.listdir("Attendance/"):
                        if file.endswith(".csv"):
                            with open("Attendance/" + file, 'r') as csvFile1:
                                reader1 = csv.reader(csvFile1)
                                for lines in reader1:
                                    if len(lines) > 0 and lines[0] == student_id:
                                        report_text.insert(tk.END, f"{lines[0]} - {lines[2]} - {lines[4]} - {lines[6]}\n")

            def download_report(report):
                import os
                folder_path = os.path.join(os.getcwd(), 'Reports')
                if not os.path.exists(folder_path):
                    os.makedirs(folder_path)
                ts = time.time()
                date = datetime.datetime.fromtimestamp(ts).strftime('%d-%m-%Y')
                file_path = os.path.join(folder_path, f"{date}_report.csv")
                with open(file_path, "w") as file:
                    file.write(report)
                mess._show(title='Report Saved', message='Report saved successfully!')
        else:
            mess._show(title='Wrong Password', message='You have entered the wrong password')

############################### USED STUFFS ####################


global key
key = ''
ts = time.time()
date = datetime.datetime.fromtimestamp(ts).strftime('%d-%m-%Y')
day,month,year=date.split("-")
mont={'01':'January',
 '02':'February',
 '03':'March',
 '04':'April',
 '05':'May',
 '06':'June',
 '07':'July',
 '08':'August',
 '09':'September',
 '10':'October',
 '11':'November',
 '12':'December'
}


############################## GUI ############################

window = tk.Tk()
window.geometry("1920x1080")
window.resizable(True, False)
window.title("AttendAI : By Govt. Ramchandra Khaitan Polytechnic College, Jaipur")

window.configure(background='#f0f0f0')

frame1 = tk.Frame(window, bg="#cce7ff")  # Light blue for a calm effect
frame1.place(relx=0.11, rely=0.12, relwidth=0.39, relheight=0.55)
frame2 = tk.Frame(window, bg="#d0f0c0")  # Light green for a fresh look
frame2.place(relx=0.51, rely=0.12, relwidth=0.39, relheight=0.55)


message3 = tk.Label(window, text="Face Recognition Based Attendance System", fg="white", bg="#003366", width=65, height=2, font=('times', 29, 'bold'))
message3.place(x=10, y=10)

frame3 = tk.Frame(window, bg="#660066")
frame3.place(relx=0.86, rely=0.025, relwidth=0.09, relheight=0.07)
frame4 = tk.Frame(window, bg="#c4c6ce")
frame4.place(relx=0.78, rely=0.025, relwidth=0.10, relheight=0.07)

datef = tk.Label(frame4, text = day+"-"+mont[month]+"-"+year+"     |",fg="white",bg="#003366" ,width=55 ,height=1,font=('times', 13, ' bold '))
datef.pack(fill='both',expand=1)
clock = tk.Label(frame3,fg="white",bg="#003366" ,width=65 ,height=1,font=('times', 13, ' bold '))
clock.pack(fill='both',expand=1)
tick()



head1 = tk.Label(frame1, text="For Already Registered" ,bg="#f0f0f0",fg="black", font=('times', 17, 'bold'))
head1.place(relx=0.05, rely=0.05, relwidth=0.9, relheight=0.1)

head2 = tk.Label(frame2, text="For New Registrations",bg="#f0f0f0" , fg="black", font=('times', 17, 'bold'))# relief='solid', bd=1)
head2.place(relx=0.05, rely=0.05, relwidth=0.9, relheight=0.1)

lbl = tk.Label(frame2, text="Enter ID", width=12, height=1, fg="black", bg="#b3e5fc", font=('times', 15, 'bold'))
lbl.place(x=52, y=220)
txt = tk.Entry(frame2, width=28, fg="black", font=('times', 17, 'bold'))
txt.place(x=208, y=220)

lbl2 = tk.Label(frame2, text="Enter Name", width=12, fg="black", bg="#b3e5fc", font=('times', 15, 'bold'))
lbl2.place(x=52, y=280)
txt2 = tk.Entry(frame2, width=28, fg="black", font=('times', 17, 'bold'))
txt2.place(x=208, y=280)

message1 = tk.Label(frame2, text="NOTE:- 1)Take Images >>> 2)Save Profile", bg="white", fg="black", width=55, height=1, font=('times', 13))
message1.place(x=52, y=345)
message = tk.Label(frame2, text="", bg="#4fc3f7", fg="black", width=39, height=1, font=('times', 16, 'bold'))
message.place(x=7, y=450)

lbl3 = tk.Label(frame1, text="Today's Attendance", width=20, fg="black", height=1, font=('times', 17, 'bold'))
lbl3.place(x=175, y=115)
lbl4 = tk.Label(frame2, text="Fill the Following Details", width=20, fg="black", height=1, font=('times', 17, 'bold'))
lbl4.place(x=175, y=125)

res = 0
exists = os.path.isfile("StudentDetails/StudentDetails.csv")
if exists:
    registered_students = []
    with open("StudentDetails/StudentDetails.csv", 'r') as csvFile1:
        reader1 = csv.reader(csvFile1)
        lst = list(reader1)
        i = 1
        for row in lst:
            i = i + 1
            if i % 2 == 0:
                registered_students.append(row[2])
    csvFile1.close()
    res = len(registered_students) - 1
else:
    res = 0

message.configure(text='Total Registrations till now : ' + str(res), width="30", bg="#d0f0c0")
message.place(x=120, y=500)


####################### MENUBAR #########################

menubar = tk.Menu(window, relief='ridge')

#Help Menu
filemenu = tk.Menu(menubar, tearoff=0)
filemenu.add_command(label='Contact Us', command=contact)
filemenu.add_command(label='Exit', command=window.destroy)
menubar.add_cascade(label='Help', menu=filemenu)

# Admin Panel Menu
admin_menu = tk.Menu(menubar, tearoff=0)
admin_menu.add_command(label='Change Password', command=change_pass)
admin_menu.add_command(label='Generate Report', command=generate_report)
menubar.add_cascade(label='Admin Panel', menu=admin_menu)


################## TREEVIEW ATTENDANCE TABLE #############


tv = ttk.Treeview(frame1, height=13, columns=('name', 'date', 'time'))
tv.column('#0', width=82)
tv.column('name', width=130)
tv.column('date', width=133)
tv.column('time', width=133)
tv.grid(row=2, column=0, padx=(50, 0), pady=(170, 0), columnspan=4)
tv.heading('#0', text='ID')
tv.heading('name', text='NAME')
tv.heading('date', text='DATE')
tv.heading('time', text='TIME')
#Initially shows the attendance on tree view
load_attendance()


######################SCROLLBAR##############################


scroll = ttk.Scrollbar(frame1, orient='vertical', command=tv.yview)
scroll.grid(row=2, column=4, padx=(12, 100), pady=(170, 1), sticky='ns')
tv.configure(yscrollcommand=scroll.set)


########################### BUTTONS ###########################


clearButton = tk.Button(frame2, text="Clear", command=clear, fg="black", bg="#ffcdd2", width=10, font=('times', 11, 'bold'))
clearButton.place(x=455, y=220)
clearButton2 = tk.Button(frame2, text="Clear", command=clear2, fg="black", bg="#ffcdd2", width=10, font=('times', 11, 'bold'))
clearButton2.place(x=455, y=280)
takeImg = tk.Button(frame2, text="Take Images", command=TakeImages, fg="white", bg="#388e3c", width=18, height=1, font=('times', 15, 'bold'))
takeImg.place(x=200, y=385)
trainImg = tk.Button(frame2, text="Save Profile", command=psw, fg="white", bg="#388e3c", width=18, height=1, font=('times', 15, 'bold'))
trainImg.place(x=200, y=429)
trackImg = tk.Button(frame1, text="Take Attendance", command=TrackImages, fg="white", bg="#388e3c", width=18, height=1, font=('times', 15, 'bold'))
trackImg.place(x=200, y=468)
quitWindow = tk.Button(frame1, text="Quit", command=window.destroy, fg="white", bg="#d32f2f", width=18, height=1, font=('times', 15, 'bold'))
quitWindow.place(x=200, y=512)


############################# END #############################


window.configure(menu=menubar)
window.mainloop()
```
